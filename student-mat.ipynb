Musimy zmienić nasze kategoryczne kolumny na kolumny liczbowe. Użyjemy do tego funkcji get_dummies z odpowiednimi atrybutami

```python
# get_dummies - drop_first pomija zmienne które są skorelowane
df = pd.get_dummies(df, drop_first=True)
```

```python
# nasza zmienna objasniana
y = df.loc[:, 'G3']

# nasze zmienne objaśniające, usuwamy zmienną G1
X = df.drop(['G1', 'G3'], axis=1)

print('nasz zbiór X: {} \n nasz zbiór y: {}'.format(X.shape, y.shape))
```

```python
# nasz zbiór treningowy i testowy
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

Żeby poprawić skuteczność naszego modelu, musimy wyrzucić pare zmiennych, które poprostu są nam nie potrzebne. Univariate feature selection polega na wybieraniu najlepszych funkcji na podstawie jednowymiarowych testów statystycznych. 
SelectPercentile usuwa wszystkie "niepotrzebne" zmienne.
Obiekty te przyjmują jako dane wejściowe funkcję scoringową, która zwraca wyniki jednokwadratowe i wartości p.

```python
from sklearn.feature_selection import f_regression, SelectPercentile

# 25% najistotniejszych zmiennych
select = SelectPercentile(score_func=f_regression, percentile=25)
select.fit(X_train, y_train)

# transformacja zbiorow
X_train_sel = select.transform(X_train)
X_test_sel = select.transform(X_test)
print("X_train: {}".format(X_train.shape))
print("X_train_sel: {}".format(X_train_sel.shape))
```

```python
# wybieramy zmienne zgodnie z najwyższymi wynikami k
mask = select.get_support()

# nasze wyselekcjonowane zmienne
new_features = X_train.columns[mask]
print(new_features)
```

Regresja liniowa mimo że jest najprostsza to często potrzeba w jej przypadku podjąć wiele krokow sprawdzających zalozenia modelu ( autokorelacja, losowość składnika losowego, heteroskedastycznosc )

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score

# inicjacja regresji liniowej
lr = LinearRegression()
lr.fit(X_train_sel, y_train)
print('Nasz wynik to: {:.3f}'.format(lr.score(X_test_sel, y_test)))
print('lr.intercept_: {}'.format(lr.intercept_))
print('lr.coef_: {}'.format(lr.coef_))
```

Dla porównania dodamy Random forest regressor

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators=500, max_depth=4, bootstrap=False, n_jobs=-1)
rf.fit(X_train, y_train)
print("Features sorted by their score:")
print(sorted(zip(map(lambda x: round(x, 4), rf.feature_importances_), X.columns), 
             reverse=True))
```

```python
import re
import IPython, graphviz
from sklearn.tree import export_graphviz
def draw_tree(t, df, size=10, ratio=0.6, precision=0):
    """ Draws a representation of a random forest in IPython.

    Parameters:
    -----------
    t: The tree you wish to draw
    df: The data used to train the tree. This is used to get the names of the features.
    """
    s=export_graphviz(t, out_file=None, feature_names=df.columns, filled=True,
                      special_characters=True, rotate=True, precision=precision)
    IPython.display.display(graphviz.Source(re.sub('Tree {',
       f'Tree {{ size={size}; ratio={ratio}', s)))

draw_tree(rf.estimators_[0], X_train, precision=3)
```
